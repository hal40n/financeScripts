```python
# 三菱UFJ銀行の預金金利引き上げ前後の運用結果の計算
# 元本100万円を20年間運用したと仮定した場合の運用

# モジュールのインポート
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import sys

# フォント設定
if sys.platform.startswith('win'):
    FontPath = 'C:\\Windows\\Fonts\\meiryo.ttc'
elif sys.platform.startswith('darwin'):
    FontPath = '/System/Library/Fonts/ヒラギノ角ゴシック W4.ttc'
elif sys.platform.startswith('linux'):
    FontPath = '/usr/share/fonts/truetype/takao-gothic/TakaoExGothic.ttf'

jpfont = FontProperties(fname=FontPath)

#  引き上げ前の金利
beforeRate = 0.001
# 引き上げ後の金利
afterRate = 0.02
# 運用期間
period = 20

# 元本
principal = 1000000

# 各年における運用結果を計算
years = np.arange(1, period + 1)
beforeAmounts = principal * (1 + beforeRate) ** years
afterAmounts = principal * (1 + afterRate) ** years

# グラフを描画
plt.figure(figsize=(10, 6))
plt.plot(years.astype(int), beforeAmounts.astype(int), label='引き上げ前', marker='o')
plt.plot(years.astype(int), afterAmounts.astype(int), label='引き上げ後', marker='o')
plt.title('引き上げ前後の運用結果', fontproperties=jpfont)
plt.xlabel('年数', fontproperties=jpfont)
plt.ylabel('運用結果', fontproperties=jpfont)
plt.legend(prop=jpfont)  # フォント設定を凡例に適用
plt.grid(True)
plt.show()
```
